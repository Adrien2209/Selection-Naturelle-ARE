```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import animation
from tkinter import *
import math,random

L=[]

LARGEUR = 960
HAUTEUR = 640
RAYON = 15 # rayon de la balle
RAYON1 = 20 # rayon de la balle1
RAYON2 = 18 # rayon de la balle2
RAYON3 = 18 # rayon de la balle3
RAYON4= 12

n=10
n1i=20

n4i=20

stop=True

# position initiale
X = []
Y = []
X1 = []
Y1 = []
X2 = []
Y2 = []
X3 = []
Y3 = []
X4 = []
Y4 = []

for i in range (n1i):
    X.append(LARGEUR*random.uniform(0.1,0.9))
    Y.append(HAUTEUR*random.uniform(0.1,0.9))
    
for i in range(n):    
    X1.append(LARGEUR*random.uniform(0.1,0.9))
    Y1.append(HAUTEUR*random.uniform(0.1,0.9))
    X2.append(LARGEUR*random.uniform(0.1,0.9))
    Y2.append(HAUTEUR*random.uniform(0.1,0.9))
    
for i in range (n4i):    
    X3.append(LARGEUR*random.uniform(0.1,0.9))
    Y3.append(HAUTEUR*random.uniform(0.1,0.9))


# direction initiale aleatoire
DX = []
DY = []
DX1 = []
DY1 = []
DX2 = []
DY2 = []
DX3 = []
DY3 = []
DX4 = []
DY4 = []

for i in range(n1i):
    vitesse = random.uniform(1.8,2)*5
    angle = random.uniform(0,2*math.pi)
    DX.append(vitesse*math.cos(angle))
    DY.append(vitesse*math.sin(angle))
    
    
for i in range(n):
    vitesse1 = random.uniform(1.8,2)*5
    angle1 = random.uniform(0,2*math.pi)
    DX1.append(vitesse1*math.cos(angle1))
    DY1.append(vitesse1*math.sin(angle1))
    
    vitesse2 = random.uniform(1.8,2)*5
    angle2 = random.uniform(0,2*math.pi)
    DX2.append(vitesse2*math.cos(angle2))
    DY2.append(vitesse2*math.sin(angle2))
    
for i in range(n4i):    
    vitesse3 = random.uniform(10,15)
    angle3 = random.uniform(0,2*math.pi)
    DX3.append(vitesse3*math.cos(angle3))
    DY3.append(vitesse3*math.sin(angle3))
    
    
    
def deplacement():
    """ Deplacement de la balle """
    global X,Y,DX,DY,RAYON,LARGEUR,HAUTEUR,stop
    
    for i in range(n):
        
        n1=len(Balle)
        n2=len(Balle1)
        n3=len(Balle2)
        n4=len(Balle3)
        
        
    for i in range(n1):
        # rebond a droite
        if X[i]+RAYON+DX[i] > LARGEUR:
            X[i] = 2*(LARGEUR-RAYON)-X[i]
            DX[i] = -DX[i]
        
        # rebond a gauche
        if X[i]-RAYON+DX[i] < 0:
            X[i] = 2*RAYON-X[i]
            DX[i] = -DX[i]
            
        # rebond en bas
        if Y[i]+RAYON+DY[i] > HAUTEUR:
            Y[i] = 2*(HAUTEUR-RAYON)-Y[i]
            DY[i] = -DY[i]
        
        # rebond en haut
        if Y[i]-RAYON+DY[i] < 0:
            Y[i] = 2*RAYON-Y[i]
            DY[i] = -DY[i]
            
        X[i] = X[i]+DX[i]
        Y[i] = Y[i]+DY[i]    
    
    
    for i in range(n2):
        # rebond a droite
        if X1[i]+RAYON1+DX1[i] > LARGEUR:
            X1[i] = 2*(LARGEUR-RAYON1)-X1[i]
            DX1[i] = -DX1[i]
            
        # rebond a gauche    
        if X1[i]-RAYON1+DX1[i] < 0:
            X1[i] = 2*RAYON1-X1[i]
            DX1[i] = -DX1[i]
            
        # rebond en bas    
        if Y1[i]+RAYON1+DY1[i] > HAUTEUR:
            Y1[i] = 2*(HAUTEUR-RAYON1)-Y1[i]
            DY1[i] = -DY1[i]
            
        # rebond en haut
        if Y1[i]-RAYON1+DY1[i] < 0:
            Y1[i] = 2*RAYON1-Y1[i]
            DY1[i] = -DY1[i]  
            
        X1[i] = X1[i]+DX1[i]
        Y1[i] = Y1[i]+DY1[i]    
    
    for i in range(n3):
        # rebond a droite
        if X2[i]+RAYON+DX2[i] > LARGEUR:
            X2[i] = 2*(LARGEUR-RAYON2)-X2[i]
            DX2[i] = -DX2[i]
        
        # rebond a gauche
        if X2[i]-RAYON2+DX2[i] < 0:
            X2[i] = 2*RAYON2-X2[i]
            DX2[i] = -DX2[i]
            
        # rebond en bas    
        if Y2[i]+RAYON2+DY2[i] > HAUTEUR:
            Y2[i] = 2*(HAUTEUR-RAYON2)-Y2[i]
            DY2[i] = -DY2[i]
            
        # rebond en haut
        if Y2[i]-RAYON2+DY2[i] < 0:
            Y2[i] = 2*RAYON2-Y2[i]
            DY2[i] = -DY2[i] 
        
        X2[i] = X2[i]+DX2[i]
        Y2[i] = Y2[i]+DY2[i]
        
    
    for i in range(n4):
        # rebond a droite
        if X3[i]+RAYON3+DX3[i] > LARGEUR:
            X3[i] = 2*(LARGEUR-RAYON3)-X3[i]
            DX3[i] = -DX3[i]
        
        # rebond a gauche
        if X3[i]-RAYON3+DX3[i] < 0:
            X3[i] = 2*RAYON3-X3[i]
            DX3[i] = -DX3[i]
            
        # rebond en bas    
        if Y3[i]+RAYON3+DY3[i] > HAUTEUR:
            Y3[i] = 2*(HAUTEUR-RAYON3)-Y3[i]
            DY3[i] = -DY3[i]
   
        # rebond en haut
        if Y3[i]-RAYON3+DY3[i] < 0:
            Y3[i] = 2*RAYON3-Y3[i]
            DY3[i] = -DY3[i]
        X3[i] = X3[i]+DX3[i]
        Y3[i] = Y3[i]+DY3[i]
    
        
       
        for a in range(n2):
            for i in range(n1):
            
                if X1[a]-RAYON1<=X[i]<=X1[a]+RAYON1 and Y1[a]-RAYON1<=Y[i]<=Y1[a]+RAYON1:
                    Canevas.delete(Balle[i])
                    Balle[i]=None
                    
            for i in range(n3):
                n6=random.uniform(0,1)    
                if n6<0.01:
                    if X1[a]-RAYON1<=X2[i]<=X1[a]+RAYON1 and Y1[a]-RAYON1<=Y2[i]<=Y1[a]+RAYON1:
                        
                        X4.append(X2[i])
                        Y4.append(Y2[i])
                        if random.uniform(0,1)>0.5:
                            d='red'
                        else:
                            d='white'
                    
                        Balle4.append(Canevas.create_oval(X2[i]-RAYON4,Y2[i]-RAYON4,X2[i]+RAYON4,Y2[i]+RAYON4,width=1,fill=d))
                
                
        for a in range(n3):
            for i in range(n4):
                if random.uniform(0,1)<0.1:
                    if X2[a]-RAYON2<=X3[i]<=X2[a]+RAYON2 and Y2[a]-RAYON2<=Y3[i]<=Y2[a]+RAYON2:
                        Canevas.delete(Balle3[i])
                        Balle3[i]=None
                    
                        c=None
                        m=0
                        for p in range(n4i):
                            m=m+1
                            if c != Balle3[p]:
                                c = True
                            if c == None and m==10:
                                stop=False
                    
            
                    
                    
        
        # affichage
    for i in range(n1):
        if Balle[i]:
            Canevas.coords(Balle[i],X[i]-RAYON,Y[i]-RAYON,X[i]+RAYON,Y[i]+RAYON)
    for i in range(n2):        
        if Balle1[i]:
            Canevas.coords(Balle1[i],X1[i]-RAYON1,Y1[i]-RAYON1,X1[i]+RAYON1,Y1[i]+RAYON1)
    for i in range(n3):        
        if Balle2[i]:
            Canevas.coords(Balle2[i],X2[i]-RAYON2,Y2[i]-RAYON2,X2[i]+RAYON2,Y2[i]+RAYON2)
    for i in range(n4):        
        if Balle3[i]:
            Canevas.coords(Balle3[i],X3[i]-RAYON3,Y3[i]-RAYON3,X3[i]+RAYON3,Y3[i]+RAYON3)
        
    for j in range(len(Balle4)):
        if Balle4[j]:
            Canevas.coords(Balle4[j],X4[j]-RAYON4,Y4[j]-RAYON4,X4[j]+RAYON4,Y4[j]+RAYON4)
        vitesse4= random.uniform(1.8,2)*5
        angle4 = random.uniform(0,2*math.pi)
        DX4.append(vitesse4*math.cos(angle4))
        DY4.append(vitesse4*math.sin(angle4))
        if X4[j]+RAYON4+DX4[j] > LARGEUR:
                X4[j] = 2*(LARGEUR-RAYON4)-X4[j]
                DX4[j] = -DX4[j]
        if X4[j]-RAYON4+DX4[j] < 0:
            X4[j] = 2*RAYON4-X4[j]
            DX4[j] = -DX4[j]
        if Y4[j]+RAYON4+DY4[j] > HAUTEUR:
            Y4[j] = 2*(HAUTEUR-RAYON4)-Y4[j]
            DY4[j] = -DY4[j]
        if Y4[j]-RAYON4+DY4[j] < 0:
            Y4[j] = 2*RAYON4-Y4[j]
            DY4[j] = -DY4[j]
            
        X4[j] = X4[j]+DX4[j]/4
        Y4[j] = Y4[j]+DY4[j]/4
            
            
            
            
        for a in range(n4):
                
            if random.uniform(0,1)>0.5:
                if Balle3[a]:
                    if X3[a]-RAYON3<=X4[j]<=X3[a]+RAYON3 and Y3[a]-RAYON3<=Y4[j]<=Y3[a]+RAYON3:
                        Canevas.delete(Balle4[j])
                        Balle4[j]=None
            
            
            

    if stop==True:            
        # mise a jour toutes les 50 ms
        Mafenetre.after(50,deplacement)
    
 
# Creation de la fenetre principale
Mafenetre = Tk()
Mafenetre.title("Animation Balle")

# Creation d'un widget Canvas
Canevas = Canvas(Mafenetre,height=HAUTEUR,width=LARGEUR,bg='white')
Canevas.pack(padx=5,pady=5)

# Creation d'un objet graphique
Balle=[]
Balle1=[]
Balle2=[]
Balle3=[]
Balle4=[]
g='green'
for i in range(n1i):
    Balle.append(Canevas.create_oval(X[i]-RAYON,Y[i]-RAYON,X[i]+RAYON,Y[i]+RAYON,width=1,fill=g))
for i in range(n):    
    Balle1.append(Canevas.create_oval(X1[i]-RAYON1,Y1[i]-RAYON1,X1[i]+RAYON1,Y1[i]+RAYON1,width=1,fill='red'))
    Balle2.append(Canevas.create_oval(X2[i]-RAYON2,Y2[i]-RAYON2,X2[i]+RAYON2,Y2[i]+RAYON2,width=1,fill='white'))
for i in range(n4i):    
    Balle3.append(Canevas.create_oval(X3[i]-RAYON3,Y3[i]-RAYON3,X3[i]+RAYON3,Y3[i]+RAYON3,width=1,fill='blue'))
    
    

   

deplacement()
Mafenetre.mainloop()
```
