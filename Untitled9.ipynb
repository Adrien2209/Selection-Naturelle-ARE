```python
from tkinter import *
import math,random

LARGEUR = 960
HAUTEUR = 640
RAYON = 15 # rayon de la balle
RAYON1 = 20 # rayon de la balle1
RAYON2 = 12 # rayon de la balle2
RAYON3 = 9 # rayon de la balle3

n=25

# position initiale
X = []
Y = []
X1 = []
Y1 = []
X2 = []
Y2 = []
X3 = []
Y3 = []
for i in range(n):
    X.append(LARGEUR*random.uniform(0.1,0.9))
    Y.append(HAUTEUR*random.uniform(0.1,0.9))
    X1.append(LARGEUR*random.uniform(0.1,0.9))
    Y1.append(HAUTEUR*random.uniform(0.1,0.9))
    X2.append(LARGEUR*random.uniform(0.1,0.9))
    Y2.append(HAUTEUR*random.uniform(0.1,0.9))
    X3.append(LARGEUR*random.uniform(0.1,0.9))
    Y3.append(HAUTEUR*random.uniform(0.1,0.9))


# direction initiale aleatoire
DX = []
DY = []
DX1 = []
DY1 = []
DX2 = []
DY2 = []
DX3 = []
DY3 = []
for i in range(n):
    vitesse = random.uniform(1.8,2)*5
    angle = random.uniform(0,2*math.pi)
    DX.append(vitesse*math.cos(angle))
    DY.append(vitesse*math.sin(angle))
    vitesse1 = random.uniform(1.8,2)*5
    angle1 = random.uniform(0,2*math.pi)
    DX1.append(vitesse1*math.cos(angle1))
    DY1.append(vitesse1*math.sin(angle1))
    vitesse2 = random.uniform(1.8,2)*5
    angle2 = random.uniform(0,2*math.pi)
    DX2.append(vitesse2*math.cos(angle2))
    DY2.append(vitesse2*math.sin(angle2))
    vitesse3 = random.uniform(1.8,2)*5
    angle3 = random.uniform(0,2*math.pi)
    DX3.append(vitesse3*math.cos(angle3))
    DY3.append(vitesse3*math.sin(angle3))
def deplacement():
    """ Deplacement de la balle """
    global X,Y,DX,DY,RAYON,LARGEUR,HAUTEUR
    
    for i in range(n):
        # rebond a droite
        if X[i]+RAYON+DX[i] > LARGEUR:
            X[i] = 2*(LARGEUR-RAYON)-X[i]
            DX[i] = -DX[i]
        if X1[i]+RAYON1+DX1[i] > LARGEUR:
            X1[i] = 2*(LARGEUR-RAYON1)-X1[i]
            DX1[i] = -DX1[i]
        if X2[i]+RAYON+DX2[i] > LARGEUR:
            X2[i] = 2*(LARGEUR-RAYON2)-X2[i]
            DX2[i] = -DX2[i]
        if X3[i]+RAYON3+DX3[i] > LARGEUR:
            X3[i] = 2*(LARGEUR-RAYON3)-X3[i]
            DX3[i] = -DX3[i]
   
        # rebond a gauche
        if X[i]-RAYON+DX[i] < 0:
            X[i] = 2*RAYON-X[i]
            DX[i] = -DX[i]
        if X1[i]-RAYON1+DX1[i] < 0:
            X1[i] = 2*RAYON1-X1[i]
            DX1[i] = -DX1[i]
        if X2[i]-RAYON2+DX2[i] < 0:
            X2[i] = 2*RAYON2-X2[i]
            DX2[i] = -DX2[i]
        if X3[i]-RAYON3+DX3[i] < 0:
            X3[i] = 2*RAYON3-X3[i]
            DX3[i] = -DX3[i]
    
        # rebond en bas
        if Y[i]+RAYON+DY[i] > HAUTEUR:
            Y[i] = 2*(HAUTEUR-RAYON)-Y[i]
            DY[i] = -DY[i]
        if Y1[i]+RAYON1+DY1[i] > HAUTEUR:
            Y1[i] = 2*(HAUTEUR-RAYON1)-Y1[i]
            DY1[i] = -DY1[i]
        if Y2[i]+RAYON2+DY2[i] > HAUTEUR:
            Y2[i] = 2*(HAUTEUR-RAYON2)-Y2[i]
            DY2[i] = -DY2[i]
        if Y3[i]+RAYON3+DY3[i] > HAUTEUR:
            Y3[i] = 2*(HAUTEUR-RAYON3)-Y3[i]
            DY3[i] = -DY3[i]

        # rebond en haut
        if Y[i]-RAYON+DY[i] < 0:
            Y[i] = 2*RAYON-Y[i]
            DY[i] = -DY[i]
        if Y1[i]-RAYON1+DY1[i] < 0:
            Y1[i] = 2*RAYON1-Y1[i]
            DY1[i] = -DY1[i]
        if Y2[i]-RAYON2+DY2[i] < 0:
            Y2[i] = 2*RAYON2-Y2[i]
            DY2[i] = -DY2[i]
        if Y3[i]-RAYON3+DY3[i] < 0:
            Y3[i] = 2*RAYON3-Y3[i]
            DY3[i] = -DY3[i]
    
        X[i] = X[i]+DX[i]
        Y[i] = Y[i]+DY[i]
        X1[i] = X1[i]+DX1[i]
        Y1[i] = Y1[i]+DY1[i]
        X2[i] = X2[i]+DX2[i]
        Y2[i] = Y2[i]+DY2[i]
        X3[i] = X3[i]+DX3[i]
        Y3[i] = Y3[i]+DY3[i]
        # affichage
        Canevas.coords(Balle[i],X[i]-RAYON,Y[i]-RAYON,X[i]+RAYON,Y[i]+RAYON)
        Canevas.coords(Balle1[i],X1[i]-RAYON1,Y1[i]-RAYON1,X1[i]+RAYON1,Y1[i]+RAYON1)
        Canevas.coords(Balle2[i],X2[i]-RAYON2,Y2[i]-RAYON2,X2[i]+RAYON2,Y2[i]+RAYON2)
        Canevas.coords(Balle3[i],X3[i]-RAYON3,Y3[i]-RAYON3,X3[i]+RAYON3,Y3[i]+RAYON3)

    # mise a jour toutes les 50 ms
    Mafenetre.after(50,deplacement)

# Creation de la fenetre principale
Mafenetre = Tk()
Mafenetre.title("Animation Balle")

# Creation d'un widget Canvas
Canevas = Canvas(Mafenetre,height=HAUTEUR,width=LARGEUR,bg='white')
Canevas.pack(padx=5,pady=5)

# Creation d'un objet graphique
Balle=[]
Balle1=[]
Balle2=[]
Balle3=[]
for i in range(n):
    Balle.append(Canevas.create_oval(X[i]-RAYON,Y[i]-RAYON,X[i]+RAYON,Y[i]+RAYON,width=1,fill='green'))
    Balle1.append(Canevas.create_oval(X1[i]-RAYON1,Y1[i]-RAYON1,X1[i]+RAYON1,Y1[i]+RAYON1,width=1,fill='red'))
    Balle2.append(Canevas.create_oval(X2[i]-RAYON2,Y2[i]-RAYON2,X2[i]+RAYON2,Y2[i]+RAYON2,width=1,fill='white'))
    Balle3.append(Canevas.create_oval(X3[i]-RAYON3,Y3[i]-RAYON3,X3[i]+RAYON3,Y3[i]+RAYON3,width=1,fill='blue'))

deplacement()
Mafenetre.mainloop()
```
